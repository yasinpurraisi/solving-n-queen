#### Fitness Function


```python
def fitness(chromosome):
    num_hits=0
    for i in range(len(chromosome)-1):
        for j in range(i+1,len(chromosome)):
            if abs(i-j) == abs(chromosome[i]-chromosome[j]):
                num_hits +=1
    
    return num_hits
```

#### Selection

```python
import random
def tournament_selection(population,k=3):
    selected = random.sample(population, k)
    selected.sort(key=lambda x: x[1])
    # print(selected)
    return selected[0][0]
```

#### Crossover

```python
def crossover(p1, p2):
    n = len(p1)  
    a, b = sorted(random.sample(range(n), 2))
    child = [None] * n

    # slice from Parent1
    child[a:b] = p1[a:b]

    fill_vals = [x for x in p2 if x not in child] # values from parent 2
    fill_idx = [i for i in range(n) if child[i] is None] # empty index

    for i, val in zip(fill_idx, fill_vals):
        child[i] = val

    return child
```

#### Mutation

```python
def mutation(chrom):
    a, b = random.sample(range(n), 2)
    chrom[a], chrom[b] = chrom[b], chrom[a]
    return chrom
```

#### Initialization

```python
n = 16
POPULATION_SIZE = 100
GENERATIONS = 500
MUTATION_RATE = 0.05
population = []

import random
for _ in range(POPULATION_SIZE):
    chromosome = list(range(1, n + 1))  
    random.shuffle(chromosome)
    score = fitness(chromosome)
    population.append((chromosome,score))

population[:5]
```

#### Main Loop

```python
for gen in range(GENERATIONS):
    next_gen = []
    sorted_population = sorted(population, key=lambda x: x[1])
    next_gen.append(sorted_population[0][0]) # Elitism

    while len(next_gen) < POPULATION_SIZE:
        parent1 = tournament_selection(population)
        parent2 = tournament_selection(population)
        child = crossover(parent1, parent2)
        if random.random() < MUTATION_RATE:
            child = mutation(child)
        next_gen.append(child)
    
    # Calculate fitness for the new generation
    population = [(chrom, fitness(chrom)) for chrom in next_gen]

    # Optional: print best fitness in this generation
    best_fitness = min(score for _, score in population)
    if gen % 50 == 0:  # Print every 50 generations
        print(f"Generation {gen}: Best Fitness = {best_fitness}")

sorted_population = sorted(population, key=lambda x: x[1])
best_ans = sorted_population[0][0]
best_fitness = sorted_population[0][1]
print(f"Best solution with {best_fitness} fitness:")
print(best_ans)
```
