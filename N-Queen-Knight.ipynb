$$ğŸš© : |r1-r2|=2  \land  |c1-c2|=1$$
$$ğŸ´ : |r1-r2|=1  \land  |c1-c2|=2$$


```python
def heuristic_func(answer):
    num_hits=0
    for i in range(len(answer)-1):
        for j in range(i+1,len(answer)):
            if abs(i-j) == abs(answer[i]-answer[j]):
                num_hits +=1
            if (abs(i - j) == 2 and abs(answer[i]-answer[j]) == 1) or (abs(i - j) == 1 and abs(answer[i]-answer[j]) == 2):
                num_hits +=1

    return num_hits
```

#### Starting Position

```python
import random

n = 16 # Chess Board Dimention
max_iter = 5000

num_restart = 50
best_score = 100
best_first_state = []
for _ in range(num_restart):
    ans = list(range(1, n + 1))  
    random.shuffle(ans)
    score = heuristic_func(ans)
    if score < best_score:
        best_first_state = ans

print(score , best_first_state)
```

#### Main Loop

```python
current_state = best_first_state
best_ans = best_first_state
k = 5
found = False
for _ in range(max_iter):
    neighbors = []
    pairs = set()
    while len(pairs) < k:
        a, b = random.sample(range(0, n-1), 2)  
        pairs.add((a, b)) 

    for pair in pairs:  # find k random neighbors
        temp = current_state.copy()   
        i = pair[0]
        j = pair[1]         
        temp[i], temp[j] = temp[j], temp[i]  
        temp_score = heuristic_func(temp)
        if temp_score == 0:
            current_state = temp
            found = True
            break 
        neighbors.append((temp,temp_score))
    if found:
        break

    sorted_neighbors = sorted(neighbors, key=lambda x: x[1])
    best_neighbor = sorted_neighbors[0][0]
    current_state = best_neighbor

print(heuristic_func(current_state) , current_state)
```
