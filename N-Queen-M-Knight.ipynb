```python
def heuristic_func(answer):
    num_hits=0
    queens = answer[0]
    knights = answer[1]
    # queen vs queen & knight
    for i in range(len(queens)-1):
        for j in range(i+1,len(queens)):
            if abs(i-j) == abs(queens[i]-queens[j]):
                num_hits +=1
            if knights[j] != 0:
                if abs(i-j) == abs(queens[i]-knights[j]):
                    num_hits +=1
    # knight vs knight & queen
    for i in range(len(knights)-1):
        for j in range(i+1,len(knights)):
            if (knights[j] != 0) and (knights[i] != 0):
                if (abs(i - j) == 2 and abs(knights[i]-knights[j]) == 1) or (abs(i - j) == 1 and abs(knights[i]-knights[j]) == 2):
                    num_hits +=1
            if (knights[i] != 0):
                if (abs(i - j) == 2 and abs(knights[i]-queens[j]) == 1) or (abs(i - j) == 1 and abs(knights[i]-queens[j]) == 2):
                    num_hits +=1

    return num_hits
```

#### Starting Position

```python
import random
n = 16 # queens
m = n//2 # knights
max_iter = 5000
num_restart = 20
best_score = n*n
best_first_state = []
for _ in range(num_restart):
    queens = list(range(1, n + 1))  
    random.shuffle(queens)
    knights = [0] * n
    knights_rows = random.sample(range(1,n+1), m)
    # knights_colm = [x for x in queens if x+1 not in knights_rows]
    for row in knights_rows:
        empty = list(range(1, n + 1))
        empty.remove(queens[row-1])
        column = random.choice(empty)
        knights[row-1]=column
    
    ans = (queens,knights)
    score = heuristic_func(ans)
    if score < best_score:
        best_first_state = ans

print(best_first_state,heuristic_func(best_first_state))
```

#### Main Loop

```python
current_state = best_first_state
k = 5
for _ in range(max_iter):
    neighbors = []
    for x in range(k):
        if random.random() < 0.5: 
            i, j = random.sample(range(0, n-1), 2)  
            temp =  current_state[0].copy()       
            temp[i], temp[j] = temp[j], temp[i]  
            neighbors.append((temp,current_state[1]))
        else:
            temp =  current_state[1].copy()
            non_zero = [i for i, x in enumerate(temp) if x != 0]
            i, j = random.sample(non_zero, 2)
            temp[i], temp[j] = temp[j], temp[i]
            neighbors.append((current_state[0],temp))
    

    best_neighbor = neighbors[0] 
    best_score = heuristic_func(best_neighbor)
    
    for neighbor in neighbors[1:]:  
        score = heuristic_func(neighbor)
        if score < best_score:
            best_neighbor = neighbor
            best_score = score
    
    if best_score <= heuristic_func(current_state):
        current_state = best_neighbor


print("Final state:", current_state)
print("Final score:", heuristic_func(current_state))
```
